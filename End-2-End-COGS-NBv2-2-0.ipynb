**Purpose:** *Stage III is the section of the Notebook where your Pandas/Python skills come into play, this is where the scraped data (in our case an Excel Spreadsheet) will be transformed from human readable to machine readable format. We call this output transform 'tidy-data'.*

```python
# Get coding! This is where we code our data transforms. We currently have a requirement within the COGS team to
# produce a best practice and hints & tips guide. I personally prefer to data-wrangle using a Pandas dataframe over
# using DataBaker (M. Spooner).


str_tabsheetsinfocus = 'H' # Only process spreadsheet tabs beginning with this character.
if boo_pythonNB_env == True:
    COGSTransformPack.printmd("**Processing tabs that start with: " + str_tabsheetsinfocus + ".**")
    
i = 0 # Loop Variable.
ii = 0 # Loop Variable.
dataframe_collection = {} # Collection of Pandas DataFrames.


for tab in tabs:
    if tabs[i].name.startswith(str_tabsheetsinfocus):
        
        
        if tabs[i].name == 'HE - Country level estimates':
            ii = 1
            try:
                pd_tab = distribution.as_pandas(sheet_name = tabs[i].name) #, skiprows=1, header=None)
                pd_df_dimensions = pd_tab.iloc[:, :7] # This split is lost in our transformation - but it helped here.
                pd_df_observations = pd_tab.iloc[:, 7:14] # See above.
                pd_df_original = pd.concat([pd_df_dimensions, pd_df_observations], axis=1, sort=False)
                pd_df_transformed = pd_df_original.dropna(how='all')
                pd_df_transformed = pd_df_transformed.drop(columns=['Country', 'sex1', 'ageband'])
                pd_df_transformed['Period'] = pd_df_transformed['Period']\
                    .map(lambda x: f'gregorian-interval/{str(x)[:4]}-03-31T00:00:00/P3Y')
                pd_df_transformed.loc[pd_df_transformed['Sex'] == 'Males', 'Sex'] = 'M'
                pd_df_transformed.loc[pd_df_transformed['Sex'] == 'Females', 'Sex'] = 'F'
                pd_df_transformed['age group'][pd_df_transformed['age group'] == '<1'] = 'lessthan1'
                pd_df_transformed['age group'][pd_df_transformed['age group'] == '90+'] = '90plus'
                pd_df_transformed_le =\
                    pd_df_transformed[['Period', 'Code', 'Sex', 'age group', 'Life Expectancy (LE)_',\
                                       'LE Lower CI_', 'LE Upper CI_',\
                                       #'Proportion of Life Spent in "Good" Health (%)_']].copy()
                                      ]].copy()
                pd_df_transformed_hle =\
                    pd_df_transformed[['Period', 'Code', 'Sex', 'age group', 'Healthy Life Expectancy (HLE) _',\
                                       'HLE Lower CI_', 'HLE Upper CI_',\
                                       #'Proportion of Life Spent in "Good" Health (%)_']].copy()
                                       ]].copy()
                pd_df_transformed_le['TransformationType'] = 'LE'
                pd_df_transformed_hle['TransformationType'] = 'HLE'
                dataframe_collection[tabs[i].name + '_LE'] = pd_df_transformed_le
                dataframe_collection[tabs[i].name + '_HLE'] = pd_df_transformed_hle
                COGSTransformPack.printmd('[' + str(i) + '] Processed: ' + tabs[i].name + '.', colour='Green')
            except ERR_HECountryLevelEstimates:
                print('Error within ' + str(pd_df_name.append(tabs[i].name)) + ' process to extract to Pandas DF.')
            
            
        if tabs[i].name == 'HE - Region level estimates':
            ii = 1
            try:
                pd_tab = distribution.as_pandas(sheet_name = tabs[i].name)
                pd_df_dimensions = pd_tab.iloc[:, :8]
                pd_df_observations = pd_tab.iloc[:, 8:14]
                pd_df_original = pd.concat([pd_df_dimensions, pd_df_observations], axis=1, sort=False)
                pd_df_original.columns = pd_df_original.iloc[0]
                pd_df_original = pd_df_original[1:]
                pd_df_transformed = pd_df_original.dropna(how='all')
                pd_df_transformed = pd_df_transformed.drop(columns=['Area_name', 'sex1', 'ageband'])
                pd_df_transformed['Period'] = pd_df_transformed['Period']\
                    .map(lambda x: f'gregorian-interval/{str(x)[:4]}-03-31T00:00:00/P3Y')
                pd_df_transformed.loc[pd_df_transformed['Sex'] == 'Males', 'Sex'] = 'M'
                pd_df_transformed.loc[pd_df_transformed['Sex'] == 'Females', 'Sex'] = 'F'
                pd_df_transformed['Age group'][pd_df_transformed['Age group'] == '<1'] = 'lessthan1'
                pd_df_transformed['Age group'][pd_df_transformed['Age group'] == '90+'] = '90plus'
                pd_df_transformed_le =\
                    pd_df_transformed[['Period', 'Code', 'Sex', 'Age group', 'Life Expectancy (LE)_',\
                                       'LE Lower CI_', 'LE Upper CI_',\
                                       #'Proportion of Life Spent in "Good" Health (%)_']].copy()
                                       ]].copy()
                pd_df_transformed_hle =\
                    pd_df_transformed[['Period', 'Code', 'Sex', 'Age group', 'Healthy Life Expectancy (HLE) _',\
                                       'HLE Lower CI_', 'HLE Upper CI_',\
                                       #'Proportion of Life Spent in "Good" Health (%)_']].copy()
                                        ]].copy()
                pd_df_transformed_le['TransformationType'] = 'LE'
                pd_df_transformed_hle['TransformationType'] = 'HLE'
                dataframe_collection[tabs[i].name + '_LE'] = pd_df_transformed_le
                dataframe_collection[tabs[i].name + '_HLE'] = pd_df_transformed_hle
                COGSTransformPack.printmd('[' + str(i) + '] Processed: ' + tabs[i].name + '.', colour='Green')
            except ERR_HERegionLevelEstimates:
                print('Error within ' + str(pd_df_name.append(tabs[i].name)) + ' process to extract to Pandas DF.')            
    
        if ii == 0:
                COGSTransformPack.printmd('[' + str(i) + '] Ignoring: ' + tabs[i].name + '.', colour='Red')
        
    i += 1
    ii = 0 # Code should be amended to utilise loop break outs...

```

```python
# ?????????????????????????????????????????????.
# Dependencies:
def __get_synonym_REFData_candidates__(dataframe_mapped_elements_collection):

    # Setup:
    df_components_from_collection = pd.DataFrame(columns=['component'])
    df_components_from_collection.iloc[0:0]
    df_components_with_synonyms = pd.DataFrame(columns=['component', 'synonym', 'synonym_score'])
    lst_components_from_collection = []

    for key in dataframe_mapped_elements_collection.keys():
        for idx_col in dataframe_mapped_elements_collection[key].index:
            lst_components_from_collection.append(dataframe_mapped_elements_collection[key][key].loc[idx_col])
    lst_components_from_collection = list(dict.fromkeys(lst_components_from_collection))
    lst_components_from_collection = list(filter(None, lst_components_from_collection))

    
    for item in lst_components_from_collection:
        df_components_from_collection = df_components_from_collection.append({'component' : item} , ignore_index=True)

    for idx_clx in df_components_from_collection.index:
        # Setup:
        lst_for_synonyms_built = []
        # Get RAW component / label / name:
        str_element = (df_components_from_collection['component'].loc[idx_clx])
        # Get Spaced [ ] component / label / name:
        lst_for_synonyms_spc = (df_components_from_collection['component'].loc[idx_clx]).split(' ')
        # Get Underscored [_] component / label / name:
        lst_for_synonyms_uds = (df_components_from_collection['component'].loc[idx_clx]).split('_')
        # Get CamelCased component / label / name:
        lst_for_synonyms = re.findall(r'[A-Z]?[a-z]+|[A-Z]+(?=[A-Z]|$)', df_components_from_collection['component'].loc[idx_clx]) # This is a CamelCase Filter (could be refactored / revised?)
        lst_for_synonyms_built = lst_for_synonyms_built + lst_for_synonyms_spc + lst_for_synonyms_uds + lst_for_synonyms
        lst_for_synonyms_built.append(str(str_element)) # Add into list original component label...
 
        # Unique and Populated:
        lst_for_synonyms_built = list(dict.fromkeys(lst_for_synonyms_built))
        lst_for_synonyms_built = list(filter(None, lst_for_synonyms_built))
        lst_for_synonyms_built = list(filter(lambda myItem: len (myItem) > 2, lst_for_synonyms_built))
        
        # Apply synonyms:
        lst_synonyms = []
        for i in lst_for_synonyms_built:    
            for syn in wordnet.synsets(str(i)):
                for lm in syn.lemmas():
                    lst_synonyms.append(lm.name())
        lst_synonyms_complete = []
        lst_synonyms_complete = lst_synonyms + lst_for_synonyms_built
        lst_synonyms_complete = list(dict.fromkeys(lst_synonyms_complete))
        lst_synonyms_complete = list(filter(None, lst_synonyms_complete))
        
        # Build DataFrame for Return:
        for x in lst_synonyms_complete:
            df_components_with_synonyms = df_components_with_synonyms.append({'component' : str_element, 'synonym' : str(x), 'synonym_score' : fuzz.token_sort_ratio(str_element, str(x))} , ignore_index=True)
            
    return df_components_with_synonyms
```

```python
# Prepares a copy of the data elements to modify. We need to keep the original data elements for later processing.
# Remember: we're passing collections of DataFrames (which hold the data elements).
# Purpose - useability of Notebooks - reduces loops and complexity.
# The REFColumnsCSV_Link was originally in place as I presumed incorrectly we were matching with Slugized Elements.
# However it is now in place to maintain lineage of the original component name and to hold the synonyms.
# Dependencies:
def __get_mapped_elements_link_key__(dataframe_elements_collection):
        
    dataframe_mapped_elements_collection = {}
    # Prepare mapping in memory component:
    for key in dataframe_elements_collection.keys(): 
        dataframe_mapped_elements_collection[key] = dataframe_elements_collection[key].copy()
        for cols in dataframe_mapped_elements_collection[key][key]:
            idx_temp = dataframe_mapped_elements_collection[key].index[dataframe_mapped_elements_collection[key][key] == cols]
            dataframe_mapped_elements_collection[key].loc[idx_temp, 'REFColumnsCSV Link'] = cols #slug.slug(cols) Not matched on Slugilzed as first believed!
            
    return dataframe_mapped_elements_collection
```
